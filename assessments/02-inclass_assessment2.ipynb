Exercise
You are given two lists of tuples representing social connections between two players from an online videogame platform. Each tuple has three elements:

player_id_1 (int)

player_id_2 (int)

date (string in MM/DD/YYYY)

Friendships: (p1, p2, start_date) — the date a friendship was established.

Unfriendships: (p1, p2, end_date) — the date the same pair unfriended.

Important: Player ID order is not guaranteed. The pairs (100, 101) and (101, 100) refer to the same two players.

Task
Produce a merged list where each record contains: (player_min_id, player_max_id, start_date, end_date_or_ongoing)

Use the same player pair key regardless of order (always sorted ascending in output).

If an unfriend exists for a pair, include its end date.

If no unfriend exists for a pair, set the end field to the string "ongoing".

Ignore unfriend records without a matching friendship.

Input Format
Two Python lists:

friendship: List[Tuple[int, int, str]]

unfriendship: List[Tuple[int, int, str]]

Dates are strings in MM/DD/YYYY.

Example Input
friendship = [
    (100, 101, "03/22/2025"),
    (100, 105, "03/18/2024")
]

unfriendship = [
    (101, 100, "03/23/2025"),  # same pair as (100, 101)
    (105, 106, "02/22/2024")   # no matching friendship -> ignore
]
Expected Output
A Python list of tuples: List[Tuple[int, int, str, Union[str, Literal["ongoing"]]]]

[
    (100, 101, "03/22/2025", "03/23/2025"),
    (100, 105, "03/18/2024", "ongoing")
]

```python
friendship = [
    (100, 101, "03/22/2025"),
    (100, 105, "03/18/2024")
]
```

```python
unfriendship = [
    (101, 100, "03/23/2025"),  # same pair as (100, 101)
    (105, 106, "02/22/2024")   # no matching friendship -> ignore
]
```

```python
friendship_dict = {}

for p1, p2, start_date in friendship:
    key = tuple(sorted((p1, p2)))
    friendship_dict[key] = {"start_date":start_date , "end_date":"ongoing"}
```

```python
friendship_dict
```

```python
for p1, p2, end_date in unfriendship:
    key = tuple(sorted((p1, p2)))
    if key in friendship_dict:
        friendship_dict[key]["end_date"] = end_date
```

```python
friendship_dict
```

```python
results = list(friendship_dict.items())
```

```python
results
```

```python
results = [] 
for (p1, p2), dates in friendship_dict.items():
    results.append((p1, p2, dates["start_date"], dates["end_date"]))
```
